```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Output/data.csv', sep=';')
```

## README + небольшие выводы:
- Я попытался **минимизировать** количество выбросов с помощью увеличения количества тестов для усреднения времени выполнения алгоритма, но, к сожалению, на некоторых графиках они все равно присутствуют, несмотря на то, что каждый алгоритм прорабатывался на одних данных **ровно 100 раз**.
- На удивление **КМП с уточненными гранями** почти всегда **требует большего времени выполнения**, чем обычный **алгоритм КМП**. Что касается **наивного алгоритма**, то он как **проигрывает**, так и **выигрывает** по времени у **КМП с уточненными гранями**, однако что можно явно отметить - **наивный алгоритм хуже всех справляется с текстом длиной 100000 символов**.
- **Алгоритм КМП**, обычно, **выигрывает по времени** у остальных двух рассматриваемых.

## Без символов подстановки, 10000 символов, бинарный алфавит

```python
bin_10000 = data[(data['Wildcards'] == 0) & (data['Text size'] == 10000) & (data['Alphabet size'] == 2)]
algorithms = bin_10000['Method'].unique()
for algorithm in algorithms:
    algo = bin_10000[bin_10000['Method'] == algorithm]
    plt.plot(algo['Pattern size'], algo['Time'], label=algorithm)

plt.title('Без символов подстановки, 10000 символов, бинарный алфавит')
plt.xlabel('Размер паттерна')
plt.ylabel('Время, нс.')
plt.legend()
plt.show()
```
